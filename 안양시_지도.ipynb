## 파일 정리, 시각화

```python
with open('anyang.geojson') as response:
  geo_json = json.load(response)
```

```python
df_num = pd.read_csv('안양시_2022_2023_인구변화.csv')
```

```python
vworld_key="4034FAEE-737C-3416-A496-E4E7FFA4D26E"

# 배경지도 타일 설정하기
layer = "Base"
tileType = "png"
tiles = f"http://api.vworld.kr/req/wmts/1.0.0/{vworld_key}/{layer}/{{z}}/{{y}}/{{x}}.{tileType}"
attr = "Vworld"

m = folium.Map([37.3942527, 126.9568209,],zoom_start = 13)

folium.TileLayer(tiles = tiles, attr = attr, overlay = True, control = True).add_to(m) # 지도 이쁘게

choropleth = folium.Choropleth(
    geo_data=geo_json, # 지도 데이터
    name = 'Anyang', # 그냥 이름
    data=df_num, # 데이터
    columns = ('행정구역(동읍면)별', 'change'), # 사용되는 열
    key_on='feature.properties.ADM_NM', # 나눠지는 기준
    fill_color='RdYlBu', # 색깔
    fill_opacity = 0.7, # 색칠되는곳 투명도
    line_opacity = 0.5, # 테두리 투명도
    legend_name='안양 동별 인구 등락' # 우측위 범례
).add_to(m)

m
```

```python
m.save('안양전체등락.html')
```

## 나이대별 변화 지도 시각화

```python
ages = pd.read_csv('안양시_2022_2023_나이대별_인구변화.csv')
ages = ages.iloc[:,1:]
```

```python
ages.head()
```

```python
ages.columns
```

```python
vworld_key="4034FAEE-737C-3416-A496-E4E7FFA4D26E"

# 배경지도 타일 설정하기
layer = "Base"
tileType = "png"
tiles = f"http://api.vworld.kr/req/wmts/1.0.0/{vworld_key}/{layer}/{{z}}/{{y}}/{{x}}.{tileType}"
attr = "Vworld"

m = folium.Map([37.3942527, 126.9568209,],zoom_start = 13, tiles = tiles, attr = attr, overlay = True, control = True)

# folium.TileLayer(tiles = tiles, attr = attr, overlay = True, control = True).add_to(m) # 지도 이쁘게

for col in ages.columns[1:]:
  c = folium.Choropleth(
      geo_data=geo_json, # 지도 데이터
      name = col, # 그냥 이름
      data=ages, # 데이터
      columns = ('행정구역(동읍면)별', col), # 사용되는 열
      key_on='feature.properties.ADM_NM', # 나눠지는 기준
      fill_color='RdYlBu', # 색깔
      fill_opacity = 0.7, # 색칠되는곳 투명도
      line_opacity = 0.5, # 테두리 투명도
      legend_name= col, # 우측위 범례
      show = False # 지도 시작시 아무것도 없이
  )#.add_to(m)
  # legend 삭제 코드
  for key in c._children:
    if key.startswith('color_map'):
        del(c._children[key])

  c.add_to(m)

folium.LayerControl(collapsed=False, overlay = False, control = True).add_to(m)

m
```
